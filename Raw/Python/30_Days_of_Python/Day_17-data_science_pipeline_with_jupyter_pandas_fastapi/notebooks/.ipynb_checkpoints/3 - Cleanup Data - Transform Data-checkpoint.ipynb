```python
import os
import pandas as pd
```

```python
# current working directory
cwd = os.getcwd() # dirname(abspath(__file__))
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
working_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
```

```python
df = pd.read_csv(working_file)
```

```python
df.head(n=1200)
```

```python
df['Rank'] = -1
```

```python
df.head()
```

```python
df['Domestic %'] = df['%']
df['Foreign %'] = df['%.1']
df.head()
```

```python
# df.drop(columns=['%','%.1'])
# or
# df.drop(['%', '%.1'], axis=1)

# df = df.drop(['%', '%.1'], axis=1)
# or
df.drop(['%', '%.1'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Worldwide'] = df['Worldwide'].replace('$', "").replace(",","")
df.head()
```

```python
to_clean_cols = ['Worldwide', 'Domestic', 'Foreign']

def convert_str_to_int(current_val):
    currency_val = current_val.replace('$', "").replace(",","")
    return currency_val

def clean_col(row):
    """
    row = Pandas Series
    dataframe = collection of Pandas Series
    """
#     print(row)
    for col in to_clean_cols:
        current_val = row[col]
        row[col] = convert_str_to_int(current_val)
    return row

df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```
